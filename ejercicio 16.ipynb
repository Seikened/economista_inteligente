```python
import yfinance as yf

data = {}
valores = ["AAPL", "MSFT", "TSLA"]

for valor in valores:
    data[valor]={"data":yf.download(valor, start="2022-09-01", end="2025-9-1", interval="1mo")}
    data[valor]["rendimiento_mensual"]= data[valor]["data"]["Close"].pct_change()
    data[valor]["promedio_aritmetico"]= data[valor]["rendimiento_mensual"].mean()
    data[valor]["promedio_anualizado"]= ((1 + data[valor]["promedio_aritmetico"])**12) - 1
```

```python
for valor in valores:
    print(f"------{valor}------")
    print("-rendimiento_mensual-")
    print(data[valor]["rendimiento_mensual"][:5]*100)
    print("-rendimiento_anualizado-")
    print(data[valor]["promedio_anualizado"][:5]*100)
    print("--")
```
